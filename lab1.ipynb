```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
# load the famous iris data
irisRaw = load_iris()
 
#Convert data to Panda package
# read iris.data into a pandas DataFrame (df), including column names
iris = pd.DataFrame(data= np.c_[irisRaw['data'], irisRaw['target']],
                     columns= irisRaw['feature_names'] + ['target'])
```

# Exercise 7: Finding min and max with indexes

```python
#Searching for maximum
iris['petal length (cm)'].max()
```

```python
#Searching for index
iris[iris['petal length (cm)'] == iris['petal length (cm)'].max()].index[0]
```

```python
#Searching for minimum
iris['petal length (cm)'].min()
```

```python
#Searching for index
iris[iris['petal length (cm)'] == iris['petal length (cm)'].min()].index[0]
```
